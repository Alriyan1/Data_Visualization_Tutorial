```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset['DESCR'])
```

```python
print(cancer_dataset['feature_names'])
```

```python
df = pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalor = StandardScaler()
```

```python
scalor.fit(df)
```

```python
df.head()
```

```python
scaled_data=scalor.transform(df)
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
data_pca = pca.fit_transform(scaled_data)
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca = PCA(n_components=3)
data_pca = pca.fit_transform(scaled_data)
data_pca
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the scatter plot
scatter = ax.scatter(data_pca[:, 0], data_pca[:, 1], data_pca[:, 2],
                     c=cancer_dataset['target'], cmap='plasma')

# Add color bar to indicate the mapping of colors to target values
colorbar = plt.colorbar(scatter)
colorbar.set_label('Target Value')

# Set labels for axes
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

# Show the plot
plt.title('3D Scatter Plot of PCA Results')
plt.show()
```
