# to_excel

```python
df
```

```python
temp_df=df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
temp_df.to_excel(r'C:\Users\farde\Downloads\batsman_runs.xlsx')
```

```python
temp_df.to_excel(r'C:\Users\farde\Downloads\batsman_runs.xlsx',sheet_name='runs')
```

```python
temp_df2=df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc="sum")
```

```python
with pd.ExcelWriter(r'C:\Users\farde\Downloads\batsman2.xlsx') as writer:
    temp_df.to_excel(writer,sheet_name='runs')
    temp_df2.to_excel(writer,sheet_name='runs2')
```
