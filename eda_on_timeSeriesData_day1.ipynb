```python
df.resample(rule='A').min()
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A').max()['Open'].plot() # yearly
```

```python
df.resample(rule='QS').max()['High'].plot() #quaterly
```

```python
df.resample(rule='BA').max() # business and frequency
```

```python
df['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df.resample(rule='M').max()['Open'].plot(kind='bar',figsize=(35,15))
```
