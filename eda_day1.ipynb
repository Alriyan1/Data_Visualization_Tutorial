```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df['Country'].value_counts()[:3].plot(kind='pie',autopct='%.2f%%')
```

most of the order from India and followed by us and uk

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
df['Aggregate rating'].value_counts().sort_index().plot(kind='bar',color=['pink','yellow','orange','green','greenyellow'])
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['#FFFFE4','red','orange','yellow','green','darkgreen'])
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['#FFFFE4','red','orange','yellow','green','darkgreen'])
```

```python
ratings['Rating color'].value_counts().plot(kind='bar',color=['orange','red','yellow','green','darkgreen','#FFFFE4'])
```
